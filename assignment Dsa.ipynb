```python
#Reverse a linked list in groups of given size
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


def reverse_linked_list_in_groups(head, k):
    current = head
    next_node = None
    prev = None
    count = 0

    # Reverse the first k nodes of the linked list
    while current and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    # Recursively reverse the remaining nodes and connect with the reversed group
    if next_node:
        head.next = reverse_linked_list_in_groups(next_node, k)

    return prev


# Function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=" -> ")
        current = current.next
    print("NULL")


# Test the function
# Example linked list: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 7 -> 8 -> NULL
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)

print("Original linked list:")
print_linked_list(head)

k = 3  # Group size

head = reverse_linked_list_in_groups(head, k)

print("\nLinked list after reversing in groups of", k)
print_linked_list(head)

```

```python
#Merge a linked list into another linked list at alternate positions.
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


def merge_linked_lists(head1, head2):
    current1 = head1
    current2 = head2

    while current1 and current2:
        next1 = current1.next
        next2 = current2.next

        current1.next = current2
        current2.next = next1

        current1 = next1
        current2 = next2

    return head1


# Function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=" -> ")
        current = current.next
    print("NULL")


# Test the function
# First linked list: 1 -> 2 -> 3 -> NULL
head1 = Node(1)
head1.next = Node(2)
head1.next.next = Node(3)

# Second linked list: 4 -> 5 -> 6 -> 7 -> NULL
head2 = Node(4)
head2.next = Node(5)
head2.next.next = Node(6)
head2.next.next.next = Node(7)

print("First linked list:")
print_linked_list(head1)

print("\nSecond linked list:")
print_linked_list(head2)

head1 = merge_linked_lists(head1, head2)

print("\nMerged linked list:")
print_linked_list(head1)
```

```python
#In an array, Count Pairs with given sum
def count_pairs_with_given_sum(arr, target_sum):
    count = 0
    num_freq = {}  # Dictionary to store frequency of elements

    for num in arr:
        complement = target_sum - num
        if complement in num_freq:
            count += num_freq[complement]  # Add the frequency of complement to the count
        if num in num_freq:
            num_freq[num] += 1
        else:
            num_freq[num] = 1

    return count


# Test the function
array = [1, 5, 7, -1, 5]
target = 6

pairs_count = count_pairs_with_given_sum(array, target)

print("Number of pairs with sum", target, "in the array:", pairs_count)
```

```python
#Find duplicates in an array
def find_duplicates(arr):
    duplicates = []
    num_set = set()

    for num in arr:
        if num in num_set:
            duplicates.append(num)
        else:
            num_set.add(num)

    return duplicates


# Test the function
array = [1, 2, 3, 4, 2, 5, 6, 1, 3]
duplicate_nums = find_duplicates(array)

print("Duplicates in the array:", duplicate_nums)
```

```python
#Find the Kth largest and Kth smallest number in an array
def find_kth_largest_smallest(arr, k):
    # Sort the array in ascending order to find the Kth smallest element
    sorted_arr = sorted(arr)
    kth_smallest = sorted_arr[k - 1]

    # Sort the array in descending order to find the Kth largest element
    sorted_arr = sorted(arr, reverse=True)
    kth_largest = sorted_arr[k - 1]

    return kth_largest, kth_smallest


# Test the function
array = [9, 4, 7, 2, 1, 8, 5, 3, 6]
k = 3

kth_largest, kth_smallest = find_kth_largest_smallest(array, k)

print("Kth smallest element:", kth_smallest)
print("Kth largest element:", kth_largest)
```

```python
#Move all the negative elements to one side of the array
def move_negatives_to_one_side(arr):
    left = 0
    right = len(arr) - 1

    while left <= right:
        if arr[left] < 0 and arr[right] < 0:
            left += 1
        elif arr[left] >= 0 and arr[right] < 0:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
        elif arr[left] >= 0 and arr[right] >= 0:
            right -= 1
        else:
            left += 1
            right -= 1

    return arr


# Test the function
array = [4, -6, 3, -8, -2, 9, 10, -1]
negatives_moved = move_negatives_to_one_side(array)

print("Array with negatives moved to one side:", negatives_moved)


```

```python
#Reverse a string using a stack data structure
def reverse_string(string):
    stack = []
    reversed_string = ""

    # Push each character onto the stack
    for char in string:
        stack.append(char)

    # Pop each character from the stack and build the reversed string
    while stack:
        reversed_string += stack.pop()

    return reversed_string
sample_string = "Hello, World!"
reversed_string = reverse_string(sample_string)
print(reversed_string)

```

```python
#Evaluate a postfix expression using stack
def evaluate_postfix(expression):
    stack = []

    for char in expression:
        if char.isdigit():
            stack.append(int(char))
        else:
            operand2 = stack.pop()
            operand1 = stack.pop()

            if char == "+":
                result = operand1 + operand2
            elif char == "-":
                result = operand1 - operand2
            elif char == "*":
                result = operand1 * operand2
            elif char == "/":
                result = operand1 / operand2

            stack.append(result)

    return stack.pop()


# Test the function
postfix_exp = "34+2*"
result = evaluate_postfix(postfix_exp)

print("Postfix expression:", postfix_exp)
print("Result:", result)
```

```python
#Implement a queue using the stack data structure
class Queue:
    def __init__(self):
        self.stack_enqueue = []
        self.stack_dequeue = []

    def enqueue(self, value):
        self.stack_enqueue.append(value)

    def dequeue(self):
        if not self.stack_dequeue:
            if not self.stack_enqueue:
                return None
            while self.stack_enqueue:
                self.stack_dequeue.append(self.stack_enqueue.pop())
        return self.stack_dequeue.pop()


# Test the implementation
queue = Queue()

queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

print("Dequeued element:", queue.dequeue())  # Dequeue 1
print("Dequeued element:", queue.dequeue())  # Dequeue 2

queue.enqueue(4)

print("Dequeued element:", queue.dequeue())  # Dequeue 3
print("Dequeued element:", queue.dequeue())  # Dequeue 4
```

```python
#Delete the elements in an linked list whose sum is equal to zero
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


def delete_zero_sum_nodes(head):
    # Create a dummy node to handle cases where the head itself needs to be deleted
    dummy = Node(0)
    dummy.next = head

    current = dummy
    prefix_sum = 0
    prefix_sum_map = {}

    while current:
        prefix_sum += current.value

        if prefix_sum in prefix_sum_map:
            # Delete nodes from prefix_sum_map[prefix_sum].next to current (inclusive)
            prev = prefix_sum_map[prefix_sum].next
            prefix_sum_to_remove = prefix_sum + prev.value

            while prev != current:
                prefix_sum_to_remove += prev.value
                del prefix_sum_map[prefix_sum_to_remove]
                prev = prev.next

            # Adjust the next pointer of the previous node
            prefix_sum_map[prefix_sum].next = current.next

        else:
            prefix_sum_map[prefix_sum] = current

        current = current.next

    return dummy.next


# Test the function
# Example linked list: 3 -> 4 -> -7 -> 5 -> -6 -> 6 -> 1 -> 2 -> NULL
head = Node(3)
head.next = Node(4)
head.next.next = Node(-7)
head.next.next.next = Node(5)
head.next.next.next.next = Node(-6)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(1)
head.next.next.next.next.next.next.next = Node(2)

print("Original linked list:")
current = head
while current:
    print(current.value, end=" -> ")
    current = current.next
print("NULL")

head = delete_zero_sum_nodes(head)

print("\nLinked list after deleting nodes with zero sum:")
current = head
while current:
    print(current.value, end=" -> ")
    current = current.next
print("NULL")
```
